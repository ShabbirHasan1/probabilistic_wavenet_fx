```python
import pandas as pd
import numpy as np
from datetime import datetime
from scipy.stats import spearmanr
```

```python
# build sampler functions
# instead of lambda function use normal function to have the same column name
def bar_len(x): return len(x)

def bar_quantile_25(x): return x.quantile(0.25)
def bar_quantile_75(x): return x.quantile(0.75)
def bar_spearman(x): 
    corr, p_value = spearmanr(x, np.arange(len(x)))
    return corr
```

```python
def tick_resampler(df, rule, base):
    # time column to index
    df['ctime']  = pd.to_datetime(df['Timestamp'])
    df = df.set_index('ctime')
    df = df.drop(columns=['Timestamp'])

    # instantiate 5 min resampler object
    rs = df.resample(rule=rule, base=base)

    # sample the 5 min dataset with the following columns:
    df = rs.agg({ 'Ask price':    ['first',           # open
                                        'max',              # high
                                        'min',              # low
                                        'last',             # close
                                        'mean',             # mean
                                        'std',              # standard deviation
                                        bar_len,            # number of ticks
                                        bar_quantile_25,    # Q1
                                        bar_quantile_75,    # Q3
                                        bar_spearman]})     # Spearman correlation
    # simplify dataset as we use only the ask price                                  
    df = df['Ask price']

    # logreturn of the bar
    df['bar_log_r'] = np.log(df['last']/df['first']) # log return

    # make return-like features '_r'
    df_aux = np.log(df[['first', 'max', 'min', 'last', 'bar_quantile_25', 'bar_quantile_75']].div(df['mean'], axis=0))
    df_aux = df_aux.rename(columns={  'first':'first_r',
                                'max':'max_r', 
                                'min':'min_r', 
                                'last':'last_r', 
                                'bar_quantile_25':'bar_quantile_25_r', 
                                'bar_quantile_75':'bar_quantile_75_r'})
    df = pd.concat([df, df_aux], axis=1, join='inner')

    return df
```

```python
reverse_tick = ['USDJPY_2016_tick.csv',
                'USDJPY_2017_tick.csv',
                'USDJPY_2018_tick.csv',
                'USDJPY_2019_tick.csv']
```

```python
# reverse USDJPY ask price to JPYUSD * 100
for i in range(len(reverse_tick)):
    df = pd.read_csv(f'fx_data/{reverse_tick[i]}', sep=",")
    df['Ask price'] = 100 / df['Ask price']
    df.to_csv(f'fx_data/JPYUSD100_{reverse_tick[i].split("USDJPY_")[1]}', index=False)
```

```python
# files
tick_data = [   #'EURUSD_2016_tick.csv',
                #'EURUSD_2017_tick.csv',
                #'EURUSD_2018_tick.csv',
                #'EURUSD_2019_tick.csv',
                #'GBPUSD_2016_tick.csv',
                #'GBPUSD_2017_tick.csv',
                #'GBPUSD_2018_tick.csv',
                #'GBPUSD_2019_tick.csv',
                'JPYUSD100_2016_tick.csv',
                'JPYUSD100_2017_tick.csv',
                'JPYUSD100_2018_tick.csv',
                'JPYUSD100_2019_tick.csv']
```

```python
# read/save tick data
rule='5min'
base=0
for i in range(len(tick_data)):
    df = pd.read_csv(f'fx_data/{tick_data[i]}', sep=",")
    df = tick_resampler(df, rule, base)
    df.to_csv(f'fx_data/{tick_data[i].split("_tick.csv")[0]}_{rule}_b{base}.csv', index=True)
```
